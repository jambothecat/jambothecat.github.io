```python
# each decade, use billboard_redux

# 50s - advent of billboard

# get all unique values from certain time frame 

billboard_50s_filtered = billboard_redux[(billboard_redux['chart_week'] >= '1958-08-04') & (billboard_redux['chart_week'] <= '1959-12-31')]
billboard_50s_artists = billboard_50s_filtered["performer"].unique()
billboard_50s_artists_csv = pd.DataFrame(billboard_50s_artists)
billboard_50s_artists_csv.to_csv('billboard_50s.csv', index=False)

# 60s

billboard_60s_filtered = billboard_redux[(billboard_redux['chart_week'] >= '1960-01-01') & (billboard_redux['chart_week'] <= '1969-12-31')]
billboard_60s_artists = billboard_60s_filtered["performer"].unique()
billboard_60s_artists_csv = pd.DataFrame(billboard_60s_artists)
billboard_60s_artists_csv.to_csv('billboard_60s.csv', index=False)

# 70s

billboard_70s_filtered = billboard_redux[(billboard_redux['chart_week'] >= '1970-01-01') & (billboard_redux['chart_week'] <= '1979-12-31')]
billboard_70s_artists = billboard_70s_filtered["performer"].unique()
billboard_70s_artists_csv = pd.DataFrame(billboard_70s_artists)
billboard_70s_artists_csv.to_csv('billboard_70s.csv', index=False)


# 80s 

billboard_80s_filtered = billboard_redux[(billboard_redux['chart_week'] >= '1980-01-01') & (billboard_redux['chart_week'] <= '1989-12-31')]
billboard_80s_artists = billboard_80s_filtered["performer"].unique()
billboard_80s_artists_csv = pd.DataFrame(billboard_80s_artists)
billboard_80s_artists_csv.to_csv('billboard_80s.csv', index=False)

# 90s 

billboard_90s_filtered = billboard_redux[(billboard_redux['chart_week'] >= '1990-01-01') & (billboard_redux['chart_week'] <= '1999-12-31')]
billboard_90s_artists = billboard_90s_filtered["performer"].unique()
billboard_90s_artists_csv = pd.DataFrame(billboard_90s_artists)
billboard_90s_artists_csv.to_csv('billboard_90s.csv', index=False)



# 00s

billboard_00s_filtered = billboard_redux[(billboard_redux['chart_week'] >= '2000-01-01') & (billboard_redux['chart_week'] <= '2009-12-31')]
billboard_00s_artists = billboard_00s_filtered["performer"].unique()
billboard_00s_artists_csv = pd.DataFrame(billboard_00s_artists)
billboard_00s_artists_csv.to_csv('billboard_00s.csv', index=False)



# 10s 

billboard_10s_filtered = billboard_redux[(billboard_redux['chart_week'] >= '2010-01-01') & (billboard_redux['chart_week'] <= '2019-12-31')]
billboard_10s_artists = billboard_10s_filtered["performer"].unique()
billboard_10s_artists_csv = pd.DataFrame(billboard_10s_artists)
billboard_10s_artists_csv.to_csv('billboard_10s.csv', index=False)



# 20s - keep in mind this is early 2020s

billboard_20s_filtered = billboard_redux[(billboard_redux['chart_week'] >= '2020-01-01') & (billboard_redux['chart_week'] <= '2022-12-31')]
billboard_20s_artists = billboard_20s_filtered["performer"].unique()
billboard_20s_artists_csv = pd.DataFrame(billboard_20s_artists)
billboard_20s_artists_csv.to_csv('billboard_20s.csv', index=False)
```
